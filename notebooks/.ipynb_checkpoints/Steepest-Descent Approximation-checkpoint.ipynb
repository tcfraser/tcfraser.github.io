```python
import numpy as np
import matplotlib.pyplot as plt
import math
import mpmath
```

```python
i = 1j
```

```python
f_z = lambda z: (z - 1+i)**2
```

```python
mpmath.cplot(f_z, points=5000)
```

```python
def enf(f):
    return lambda x: math.e**(-f(x))
```

```python
f = lambda x: 0.05*((x-1)**2 * (x+1)**2)
```

```python
enf(f)(2)
```

```python
def plot_f(fs):
    for f in fs:
        v_f = np.vectorize(f)
        x = np.arange(-2, 2, 0.1)
        f_x = v_f(x)
        plt.plot(x, f_x)
```

```python
plot_f([f, enf(f)])
```
