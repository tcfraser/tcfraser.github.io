```python
import numpy as np
import matplotlib.pyplot as plt
import math
import mpmath
from math import factorial, pi
```

```python
resolution = 5000
```

```python
i = 1j
```

```python
compose = lambda f, g: lambda z: f(g(z))
```

```python
f_z = lambda z: 1/300*(z - (1+i))*(z + (1+i))**2
```

```python
e_f_z = lambda z: 100*mpmath.exp(-f_z(z))
```

```python
norm = mpmath.norm
exp = mpmath.exp
nexp = lambda x: exp(-x)
```

```python
mpmath.cplot(f_z, points=resolution)
```

```python
mpmath.cplot(compose(compose(exp, norm), f_z), points=resolution)
```

```python
def enf(f):
    return lambda x: math.e**(-f(x))
```

```python
f = lambda x: (x+2)*(x+1)*(3*x-1)*(x-2) + 10
```

```python
def ts_cos(x, terms):
    return sum((-1)**i * x**(2*i) / factorial(2*i) for i in range(terms))
```

```python
f = lambda x: (0.80)*(ts_cos(x*pi*0.8, 19) + 1.7 - 0.5*x)
```

```python
f = lambda x: 3445/21-(2242*x)/21+(5401*x**2)/216-(7339*x**3)/3024+(251*x**4)/3024
```

```python
enf(f)(2)
```

```python
def plot_f(fs):
    for f in fs:
        v_f = np.vectorize(f)
        x = np.arange(-2, 2, 0.01)
        f_x = v_f(x)
        plt.plot(x, f_x)
    plt.show()
```

```python
plot_f([f, compose(nexp, f)])
```
